<h1 align = center>Визуализация экспирементальных данных из <a href = "https://www.sciencedirect.com/science/article/pii/S0092867420310035" title = "Deep Mutational Scanning of SARS-CoV-2 Receptor Binding Domain Reveals Constraints on Folding and ACE2 Binding">статьи</a></h1>

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("single_mut_effects.csv")

def ax_format(ax, measuring, source_name, title, color, fontsize = 18):
    ax.hist(df[source_name], bins = 75, edgecolor = "black", color = color)
    ax.set_title(title, fontsize = fontsize + 6)
    ax.grid(True)
    
    if measuring == "binding":
        ax.set_xlabel("ACE2 binding", fontsize = fontsize)
    elif measuring == "expression":
        ax.set_xlabel("RBD expression", fontsize = fontsize)
        
    ax.set_ylabel("Number of cells", fontsize = fontsize)
```

<h2>Экспрессия мутантного RBD</h2>

```python
fig = plt.figure(figsize = [15, 10])
ax_lib1 = plt.subplot(221)
ax_format(ax_lib1, measuring = "expression", source_name = "expr_lib1", title = "Mutant library 1", color = "red")

ax_lib2 = plt.subplot(222)
ax_format(ax_lib2, measuring = "expression", source_name = "expr_lib2", title = "Mutant library 2", color = "blue")

ax_lib3 = plt.subplot(212)
ax_format(ax_lib3, measuring = "expression", source_name = "expr_avg", title = "Average", color = "magenta")

plt.tight_layout()
plt.show()
```

<h2>Связывание мутантного RBD c ACE2</h2>

```python
fig = plt.figure(figsize = [15, 10])
ax_lib1 = plt.subplot(221)
ax_format(ax_lib1, measuring = "binding", source_name = "bind_lib1", title = "Mutant library 1", color = "red")

ax_lib2 = plt.subplot(222)
ax_format(ax_lib2, measuring = "binding", source_name = "bind_lib2", title = "Mutant library 2", color = "blue")

ax_lib3 = plt.subplot(212)
ax_format(ax_lib3, measuring = "binding", source_name = "bind_avg", title = "Average", color = "magenta")

plt.tight_layout()
plt.show()
```
